```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

```python
#load inbuilt dataset
data = sns.load_dataset("titanic")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

Handling null values

```python
titanic=data.drop(columns='deck')
```

```python
titanic['age']=titanic['age'].fillna(titanic['age'].mean())
```

```python
titanic['embarked'].unique()
```

```python
titanic = pd.get_dummies(titanic, columns = ['embarked', 'embark_town'])
```

```python
titanic.isnull().sum()
```

# Statistical Data Visualization Using Seaborn

Joint Plot

```python
sns.jointplot(x='fare',y='age',data=titanic)
plt.title('Joint Plot')
```

Scatter Plot

```python
plt.scatter(titanic['age'],titanic['fare'],marker='*', alpha=0.8)  
plt.title('Scatter Plot')
plt.xlabel('age')
plt.ylabel('fare')
plt.show()
```

Box Plot

```python
sns.boxplot(x='pclass',y='age',data=titanic)
plt.title('Box Plot')
```

Distribution Plot

```python
sns.displot(x='age',data=titanic,bins=70)
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Distribution Plot')
```

Swarm Plot

```python
sns.swarmplot(x='pclass',y='age',data=titanic)
plt.title('Swarm Plot')
```

Count Plot

```python
sns.countplot(x='sex',data=titanic)
plt.title('Count Plot')
```

Heat Map

```python
tc = titanic.corr()
sns.heatmap(tc,cmap='coolwarm')
plt.title('Heat Map')
```

Factor Plot

```python
sns.factorplot('survived',data=titanic,kind='count',hue='pclass')
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Factor Plot for PClass')
```

```python
sns.factorplot('survived',data=titanic,kind='count',hue='sex')
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Factor Plot for Gender')
```

Histogram Plot to check how the price of the ticket for each passenger is distributed

```python
sns.histplot(x='fare',data=titanic)
sns.set(rc={'figure.figsize':(5,5)})
plt.title('Histpgram Plot')
```
